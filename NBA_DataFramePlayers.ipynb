```python
# Identificar y tratar valores faltantes en la columna 'college'
missing_college_count = all_players['college'].isnull().sum()

# Imprimir la cantidad de valores faltantes antes del tratamiento
print("Cantidad de valores faltantes en la columna 'college' antes del tratamiento:", missing_college_count)

# Rellenar los valores faltantes con un valor predeterminado
default_college_value = "No College"
all_players['college'].fillna(default_college_value, inplace=True)

# Imprimir la cantidad de valores faltantes despu√©s del tratamiento
missing_college_count_after = all_players['college'].isnull().sum()
```

```python
missing_college_count_after
```

> # **An√°lisis**

```python
# Identificar jugadores destacados
# Selecciona filas donde puntos ('pts'), rebotes ('reb') y asistencias ('ast') son mayores que el promedio respectivo
standout_players = all_players.loc[(all_players['pts'] > all_players['pts'].mean()) & (all_players['reb'] > all_players['reb'].mean()) & (all_players['ast'] > all_players['ast'].mean())].tail(10)
```

```python
# Seleccionar columnas espec√≠ficas
standout_players_filter = standout_players[['player_name', 'age', 'team_abbreviation', 'pts', 'reb', 'ast', 'season']]
```

```python
# Imprimimos los datos seleccionados y especificamos el filtro descendentemente
standout_players_filter.sort_values(by='pts', ascending=False)
```

```python
# Identificar el jugador m√°s antiguo y m√°s joven
oldest_player = all_players.loc[all_players['age'].idxmax()]
youngest_player = all_players.loc[all_players['age'].idxmin()]

# Mostrar la informaci√≥n de los jugadores
print("Jugador m√°s antiguo:")
print(oldest_player[['player_name', 'age', 'team_abbreviation', 'pts', 'reb', 'ast', 'season']])
print("\nJugador m√°s joven:")
print(youngest_player[['player_name', 'age', 'team_abbreviation', 'pts', 'reb', 'ast', 'season']])
```

> # **Visualizaci√≥n**

```python
import matplotlib.pyplot as plt

# Crear un gr√°fico de l√≠neas para la distribuci√≥n de edades
plt.plot(all_players['age'].value_counts().sort_index(), marker='o', linestyle='-', color='black')
plt.title('Distribuci√≥n de Edades de Jugadores')
plt.xlabel('Edad')
plt.ylabel('Cantidad')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Contar la frecuencia de cada altura y ordenar por √≠ndice
height_all_players = all_players['player_height'].value_counts().sort_index()

# Crear un gr√°fico de l√≠neas para la distribuci√≥n de alturas
plt.plot(height_all_players, marker='o', linestyle='-', color='black')
plt.title('Distribuci√≥n de Alturas de Jugadores')
plt.xlabel('Altura')
plt.ylabel('Cantidad')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Redondear la edad a n√∫meros enteros
all_players['age_rounded'] = all_players['age'].astype(int)

# Agrupar por altura y contar la frecuencia de edades
height_age_counts = all_players.groupby('player_height')['age_rounded'].value_counts().unstack().fillna(0)

# Crear un gr√°fico de barras apiladas
height_age_counts.plot(kind='bar', stacked=True, colormap='viridis')

plt.title('Distribuci√≥n de Edades por Altura de Jugadores')
plt.xlabel('Altura')
plt.ylabel('Frecuencia')
plt.legend(title='Edad', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
